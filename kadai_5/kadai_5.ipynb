```python
top_base = 10
bottom_base = 20
height = 5

def trapezoid_colc(top_base,bottom_base,height):
    sum1 = top_base + bottom_base
    sum2 = sum1 * height
    sum3 = sum2 / 2
    return sum3

colc = trapezoid_colc(10,20,5)
print(f"{colc}cm²")
```
